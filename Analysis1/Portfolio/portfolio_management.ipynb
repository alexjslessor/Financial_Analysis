```python
aph1 = pd.read_csv('APHcomp.csv', parse_dates=True, index_col=0)
ogi1 = pd.read_csv('OGIcomp.csv', parse_dates=True, index_col=0)
weed1 = pd.read_csv('WEEDcomp.csv', parse_dates=True, index_col=0)
```

```python
for columns in (aph1,ogi1,weed1):
    columns.drop(['Open','High','Low','Close','Volume'], 1, inplace=True)
```

```python
tickers = aph1.merge(ogi1, how='right', left_index=True, right_index=True)
tickers = tickers.merge(weed1, how='right', left_index=True, right_index=True)
tickers = tickers.rename(columns={'Adj Close_x':'APH_Adj_Close', 'Adj Close_y':'OGI_Adj_Close', 'Adj Close':'WEED_Adj_Close'})
tickers.head()
```

```python
corr = tickers.corr(method='pearson')
matrix = sns.heatmap(corr, vmin=-1, vmax=1, cmap=plt.cm.RdYlGn, linewidths=0.01)
```

```python
natlog_return = np.log(tickers / tickers.shift(1))
natlog_return.tail()
```

```python
natlog_return.hist(bins=100, figsize=(15,13))
plt.tight_layout()
```

```python
natlog_return.plot.kde(figsize=(15,13), subplots=True)
plt.tight_layout()
```

```python
np.random.seed(100)

simulations = 10000
total_weights = np.zeros((simulations, len(tickers.columns)))
returns = np.zeros(simulations)
volatility = np.zeros(simulations)
new_sharpe = np.zeros(simulations)

for sim in range(simulations):

    weights = np.array(np.random.random(3))
    weights = weights / np.sum(weights)
    
    total_weights[sim,:] = weights

    # The Expected Annual Portfolio Return
    returns[sim] = np.sum((natlog_return.mean() * weights) * 252)

    # Expected Risk Volatility
    volatility[sim] = np.sqrt(np.dot(weights.T, np.dot(natlog_return.cov() * 252, weights)))
    
    new_sharpe[sim] = returns[sim] / volatility[sim]
```
