```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
aph = pd.read_csv('APHcomp.csv', parse_dates=True, index_col=0)
ogi = pd.read_csv('OGIcomp.csv', parse_dates=True, index_col=0)
weed = pd.read_csv('WEEDcomp.csv', parse_dates=True, index_col=0)
```

```python
ogi.info()
```

```python
aph.info()
```

```python
weed.info()
```

```python
for column in (aph,ogi,weed):
    column['Portfolio_Return'] = column['Adj Close'] / column.ix[0]['Adj Close']
```

```python
for column,weights in zip((aph,ogi,weed),[.33,.33,.33]):
    column['Portfolio_Weightings'] = column['Portfolio_Return'] * weights
```

```python
for column in (aph,ogi,weed):
    column['Profit_Loss_in_Dollars'] = column['Portfolio_Weightings'] * 1000
```

```python
profitloss = [aph['Profit_Loss_in_Dollars'],ogi['Profit_Loss_in_Dollars'],weed['Profit_Loss_in_Dollars']]
totalvalue = pd.concat(profitloss, axis=1)
```

```python
totalvalue.columns = ['aph_P/L','ogi_P/L','weed_P/L']
```

```python
totalvalue['Daily_Portfolio_Return'] = totalvalue.sum(axis=1)
```

```python
totalvalue['Daily_Portfolio_Return'].plot(figsize=(10,8))
```

```python
totalvalue['Daily_Portfolio_Return_%'] = totalvalue['Daily_Portfolio_Return'].pct_change(1)
```

```python
# Cannabis stocks are quite profitable but time will tell after legalization what the equilibrium valuation will be due to the IFRS accounting these companies are using.
totalvalue['Daily_Portfolio_Return'].plot(kind='hist', bins=200, figsize=(10,8))
```

```python
totalvalue['Daily_Portfolio_Return'].plot(kind='kde', figsize=(10,8))
```

```python
sharpe_ratio = totalvalue['Daily_Portfolio_Return'].mean() / totalvalue['Daily_Portfolio_Return'].std()
```

```python
annual_sharpe_ratio = np.sqrt(252) * sharpe_ratio
```

```python
# A sharpe ratio of 2-3 is considered very good, this has a sharpe ratio of 12........
annual_sharpe_ratio
```
